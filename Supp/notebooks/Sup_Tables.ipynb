```python
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
```

```python
RANKS = ['class', 'order', 'family', 'genus', 'species']
```

# Sup_Table_S2

```python
auroc = pd.read_csv("../Sup_Table_S2/AUROC.csv", sep=',', index_col=0)
auroc['intermediate'] = pd.Categorical([m.split('/')[0] for m in auroc.index], categories=['no_network', '1D-CNN', 'GRU', '1D-CNN_GRU'])
auroc['pooler'] = pd.Categorical([m.split('/')[-1].split('.')[0] for m in auroc.index], categories=['mean', 'max', 'cls_token', 'convolutional_attention'])
auroc['phylum'] = [m.split('/')[-1].split('.')[1] for m in auroc.index]
auroc.head(n=2)
```

```python
auroc.groupby(['intermediate', 'pooler'])[RANKS].mean()
```

```python
aupr = pd.read_csv("../Sup_Table_S2/AUPR.csv", sep=',', index_col=0)
aupr['intermediate'] = pd.Categorical([m.split('/')[0] for m in aupr.index], categories=['no_network', '1D-CNN', 'GRU', '1D-CNN_GRU'])
aupr['pooler'] = pd.Categorical([m.split('/')[-1].split('.')[0] for m in aupr.index], categories=['mean', 'max', 'cls_token', 'convolutional_attention'])
aupr['phylum'] = [m.split('/')[-1].split('.')[1] for m in aupr.index]
aupr.head(n=2)
```

```python
aupr.groupby(['intermediate', 'pooler'])[RANKS].mean()
```

# Sup_Table_S3

```python
df = pd.read_csv("../Sup_Table_S3/Sup_Table_S3.data.csv", sep=',')
df['train'] = pd.Categorical(df['train'], categories=['wo_mcm', 'w_mcm'])
df['test'] = pd.Categorical(df['test'], categories=['wo_mcm', 'w_mcm'])
df['rank'] = pd.Categorical(df['rank'], categories=RANKS)
df.head(n=2)
```

```python
df.groupby(['train', 'test', 'rank'])['AUROC'].mean().unstack()
```

```python
df.groupby(['train', 'test', 'rank'])['AUPR'].mean().unstack()
```

# Supplementary Table S9

```python
df = pd.read_csv("../Sup_Table_S9/Sup_Table_S9.data.csv", sep=',')
df.head(n=2)
```

```python
auroc = df.groupby(['phylum', 'trial'])['AUROC'].mean()
auroc = auroc.unstack()
auroc['min'] = auroc.min(axis=1)
auroc['max'] = auroc.max(axis=1)
auroc['diff'] = auroc['max'] - auroc['min']
auroc = auroc[['min', 'max', 'diff']]
```

```python
auroc
```

```python
aupr = df.groupby(['phylum', 'trial'])['AUPR'].mean()
aupr = aupr.unstack()
aupr['min'] = aupr.min(axis=1)
aupr['max'] = aupr.max(axis=1)
aupr['diff'] = aupr['max'] - aupr['min']
aupr = aupr[['min', 'max', 'diff']]
```

```python
aupr
```
