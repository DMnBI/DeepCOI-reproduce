```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
t6 = pd.read_csv("DeepCOI-t6-k4.csv", sep=',')
t6['architecture'] = 't6-320'

t12 = pd.read_csv("DeepCOI-t12-k4.csv", sep=',')
t12['architecture'] = 't12-480'
```

```python
df = pd.concat([t6[:500], t12[:500]])
df.columns = ['Wall time', 'steps', 'training loss', 'architecture']
```

```python
ax = sns.lineplot(data=df,
    x='steps', y='training loss',
    hue='architecture',
)
ax.figure.savefig("../../Figures/Sup_Fig_S6.pdf", format='pdf', bbox_inches='tight')
```
